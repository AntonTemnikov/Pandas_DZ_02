```python
import pandas as pd

data = pd.read_csv('keywords.csv')

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}




def attach_region(row):
    row_list = row.split(' ')

    for key in geo_data:
        for value in geo_data[key]:
            if value in row_list:
                return key
    return 'undefined'

data['region'] = data['keyword'].apply(attach_region)

#data[data['region'] != 'undefined'].head(20)

#data.head()
```
