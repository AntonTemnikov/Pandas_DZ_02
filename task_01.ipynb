```python
import pandas as pd

data = pd.read_csv('ratings.csv')


def class_movie (assessment):
    if assessment <= 2:
        return 'Низкий рейтинг'
    elif assessment <=4:
        return 'Средний рейтинг'
    else:
        return 'Высокий рейтинг'
    
    
    
data['class_movie'] = data['rating'].apply(class_movie)

data.sort_values('rating', ascending = False)


```
